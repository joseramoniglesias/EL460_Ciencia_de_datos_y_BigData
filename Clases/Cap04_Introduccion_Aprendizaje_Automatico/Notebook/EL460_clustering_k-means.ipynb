# Clustering con el Método K-Means

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


%matplotlib inline
```

```python
vinos=pd.read_csv("caracteristicas de vinos.csv", 
                engine="python")

vinos
```

```python
vinos.info()
```

Quitar la columna "Vino" del análisis de características

```python
vinos_variables = vinos.drop(["Vino"],axis=1)
```

Obtenemos los Estadísticos Descriptivos de vinos

```python
vinos_variables.describe()
```

### Normalizar los valores de las características

```python
vinos_norm = (vinos_variables-vinos_variables.min()) / (vinos_variables.max()-vinos_variables.min())

vinos_norm
```

```python
vinos_norm.describe()
```

### Búsqueda de la cantidad óptima de clusters

Calculando qué tan similares son los individuos dentro de los clusters

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, max_iter =300)
    kmeans.fit(vinos_norm) #Aplicamos K-means a la base de datos
    wcss.append(kmeans.inertia_)
```

Graficamos los resultados de WCSS para formar el Codo de Jambú

```python
plt.plot(range(1,11),wcss)
plt.title("Codo de Jambú")
plt.xlabel("Número de Cluster")
plt.ylabel("WCSS") # WCSS es un indicador de qué tan similares son los individuos dentro de los clusters 
plt.show()
```

### Aplicar el método k-means a la base de datos

```python
clustering = KMeans(n_clusters = 3, max_iter = 300) #Crea el modelo
clustering.fit(vinos_norm) #Aplica el modelo a la base de datos
```

### Agregar la clasificación al archivo original

```python
vinos["KMeans_Clusters"] = clustering.labels_ #los resultados del clustering se guardan en labels_ dentro del modelo
vinos.head()
```

### Visualización de los clusters que se formaron

Aplicaremos el análisis de componentes principales para darnos una idea de como se formaron los clusters

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
pca_vinos = pca.fit_transform(vinos_norm)
pca_vinos_df = pd.DataFrame(data = pca_vinos, columns = ["Componente_1", "Componente_2"])
pca_nombres_vinos = pd.concat([pca_vinos_df, vinos[["KMeans_Clusters"]]], axis=1)

pca_nombres_vinos
```

```python
fig = plt.figure(figsize = (6,6))

ax = fig.add_subplot(1,1,1)
ax.set_xlabel("Componente 1", fontsize = 15)
ax.set_ylabel("Componente 2", fontsize = 15)
ax.set_title("Componentes Principales", fontsize = 15)

color_theme = np.array(["blue", "green", "orange"])
ax.scatter(x = pca_nombres_vinos.Componente_1, y = pca_nombres_vinos.Componente_2,
          c = color_theme[pca_nombres_vinos.KMeans_Clusters], s = 50)

plt.show()
```
